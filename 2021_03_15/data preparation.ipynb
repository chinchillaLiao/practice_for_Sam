```python
import shapefile # pip isntall pyshp
```

```python
sf = shapefile.Reader("../mapdata202010230200/TOWN_MOI_1091016.shp", encoding = 'utf-8')
len(sf)
```

```python
town_list = []
for x in sf:
    town_list.append(x.__geo_interface__)
town_list[0]
```

```python
d_menu = dict()
for x in town_list:
    pp = x['properties']
    countyname = pp['COUNTYNAME']
    towncode = pp['TOWNCODE']
    townname = pp['TOWNNAME']
    
    d_menu_county = d_menu.get(countyname)
    if d_menu_county:
        d_menu_county[townname] = towncode
    else:
        d_menu_county = dict()
        d_menu_county[townname] = towncode
        d_menu[countyname] = d_menu_county
    

d_menu
```

```python
d_county_gjson = dict()
for x in town_list:
    towncode = pp['TOWNCODE']
    gjson = x['geometry']
    d_county_gjson[towncode] = gjson
    
d_county_gjson
```
